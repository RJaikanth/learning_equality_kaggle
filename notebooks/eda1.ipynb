```python
import pandas as pd
```

```python
content_df = pd.read_csv("../data/content.csv")
content_df
```

```python
content_df.info()
```

```python
corr_df = pd.read_csv("../data/correlations.csv")
corr_df['content_ids'] = corr_df['content_ids'].str.split(" ")
corr_df = corr_df.explode('content_ids').rename(columns={'content_ids': 'content_id'})
corr_df
```

```python
topics_df = pd.read_csv("../data/topics.csv")
topics_df
```

```python
df = corr_df.merge(topics_df, how='left', left_on='topic_id', right_on='id')
df
```

```python
df = df[[x for x in df.columns if x != 'id']].merge(content_df, how='left', left_on='content_id', right_on='id')
df = df[[x for x in df.columns if x != 'id']]
df
```
